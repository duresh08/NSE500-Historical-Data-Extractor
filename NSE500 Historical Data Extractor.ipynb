```python
import pandas as pd
url = 'https://archives.nseindia.com/content/indices/ind_nifty500list.csv'
NSE500_data = pd.read_csv(url)
NSE500_symbols = NSE500_data['Symbol']
NSE500_symbols
```

```python
j = 0
Corrected_symbol_list = []

while j in range(len(NSE500_symbols)):
    Corrected_symbol = NSE500_symbols[j].replace('-','_')
    Corrected_symbol = Corrected_symbol.replace('&','_')
    Corrected_symbol_list.append(Corrected_symbol)
    j+=1
    
Corrected_symbol_list
```

```python
from tvDatafeed import TvDatafeed, Interval
import datetime

import pandas as pd

tv = TvDatafeed()

final_dataframe = pd.DataFrame()
i = 0

start_date = '15'
start_month = '06'
start_year = '2022'

end_date = '01'
end_month = '07'
end_year = '2022'

full_start_date = datetime.date(int(start_year),int(start_month),int(start_date))
full_end_date = datetime.date(int(end_year),int(end_month),int(end_date))

for symbols in Corrected_symbol_list:
    data = tv.get_hist(symbol='{}'.format(symbols),exchange='NSE',n_bars=5000)
    data = data[data.index.date>=full_start_date]
    data = data[data.index.date<=full_end_date]
    data = data.reset_index()
    data = data.drop("symbol",axis = 1)
    data.columns = [f'{symbols} DATE',f'{symbols} OPEN',f'{symbols} HIGH',f'{symbols} LOW',f'{symbols} CLOSE',
                    f'{symbols} VOLUME']
    final_dataframe = pd.concat([final_dataframe,data],axis = 1)
    i+=1
    
final_dataframe
```
